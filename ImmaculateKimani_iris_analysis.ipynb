# Iris Dataset Analysis

## Objective 
- Load and explore a dataset using pandas
- Perfom basic data analysis
- Create visualizations using matplotlib and seaborn
- Share findings and observations 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
try:
    from sklearn.datasets import load_iris
    iris = load_iris()
    print("Iris dataset loaded successfully.")
except FileNotFoundError:
    print("Error: The iris dataset could not be found.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Load the iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

#Display first 5 rows of the dataset
print("\nFirst 5 rows of the dataset:")
display(df.head())
```

```python
print("\nDataset Information:")
print(df.info())
```

```python
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

```python
# Summary Statistics
print("\nDescriptive Statistics:")
display(df.describe())
```

```python
grouped = df.groupby('species').mean()
print("\nMean Measurements by Species:")
display(grouped)
```

```python
# Observations
print("\nObservations:")
print("- Iris-virginica generally has the largest petal and sepal sizes.")
print("- Iris-setosa has the smallest petal length and width.")
print("- Sepal length varies less between species compared to petal measurements.")
```

```python
#1. Line Chart (using sepal length across samples to simulae trend)
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['sepal length (cm)'], label='Sepal Length', color='blue')
plt.title("Line Chart: Sepal Length Across Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()    
plt.show()
```

```python
# 2. Bar Chart (Average Petal Length by Species)
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='petal length (cm)', data=df, errorbar=None)
plt.title("Bar Chart: Average Petal Length by Species")
plt.xlabel("Species")   
plt.ylabel("Average Petal Length (cm)")
plt.show()
```

```python
# 3. Histogram (Distribution of Sepal Width)
plt.figure(figsize=(8, 6))  
plt.hist(df['sepal width (cm)'], bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram: Distribution of Sepal Width") 
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency") 
plt.show()
```

```python
# 4. Scatter Plot (Sepal Length vs Sepal Width colored by Species)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=df, palette='Set1')
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")  
plt.legend(title='Species')
plt.show()
```

## Findings & Observations
- The dataset has 150 samples and no missing values.
- Setosa species is clearly distinct due to its small petal length and width.
- Virginica species shows the largest measurements overall.
- Histograms reveal normal-like distribution for some features.
- Scatter plots confirm that petal length and width are good indicators for class separation.
